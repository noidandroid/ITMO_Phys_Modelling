```python
import math
import matplotlib.pyplot as plt
from IPython.display import clear_output
import matplotlib.animation as animation
import sys, os
```

```python
Mearth = 5.972e+21 # Earth weight


Mr = int(input("Enter Rocket Mass: "))
Mt = int(input("Enter Mass of Fuel: "))
u = int(input("Enter Gas Outflow: "))
alpha = int(input("Enter Fuel Combustion: "))
teta = int(input("Enter Takeoff Angle: "))

R = 6371 # Earth radius
g0 = 6.67e-14 * Mearth / R / R # graviation constant

x, y = 0, 0 
v = 0
i = 0
h = 0.001
m = Mt + Mr
Vx, Vy = 0, 0
plot_m, plot_v, plot_x, plot_y = [], [], [], []
plot_Vx, plot_Vy = [], []
```

```python
while y >= 0:
    g = g0 * R * R / ((R + y) * (R + y))

    if m - Mr > 0:
        Vx += math.sin(teta * math.pi / 180) * u * alpha * h / (m - alpha * h)
        Vy += math.cos(teta * math.pi / 180) * u * alpha * h / (m - alpha * h)
        m -= alpha * h

    Vy -= g * h

    if Vx != 0:
        teta = math.atan2(Vy, -Vx) * 180 / math.pi - 90

    else:
        if Vy >= 0:
            teta = 0
        else:
            teta = 180

    x += Vx * h
    y += Vy * h
    plot_m.append(m - Mr) 
    plot_v.append(math.sqrt(Vx ** 2 + Vy ** 2))
    plot_x.append(x)
    plot_y.append(y)
    plot_Vx.append(Vx)
    plot_Vy.append(Vy)

    i += 1
    if (i == 100000):
        break
```

```python
tg = [_ for _ in range(i)]

plt.plot(tg, plot_m)
plt.xlabel("t")
plt.ylabel("m(t)")
plt.show()

plt.plot(plot_x, plot_y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()

plt.plot(tg, plot_v)
plt.xlabel("t")
plt.ylabel("V(t)")
plt.show()

plt.plot(tg, plot_Vy)
plt.xlabel("t")
plt.ylabel("Vy")
plt.show()

plt.plot(tg, plot_Vx)
plt.xlabel("t")
plt.ylabel("Vx")
plt.show()
```
